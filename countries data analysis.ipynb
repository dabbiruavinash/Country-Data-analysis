```python
str_columns=['coastline_coastarea_ratio', 'net_migration',
       'infant_mortality_per_1000_births', 'literacy',
       'phones_per_1000', 'arable', 'crops', 'other', 'climate', 'birthrate',
       'deathrate', 'agriculture', 'industry', 'service']
```

```python
for column in str_columns:
    world[column] = world[column].str.replace(',','.')
    world[column] = pd.to_numeric(world[column])
```

```python
world.head()
```

```python
world['pop_density_per_sq_mi']=world['pop_density_per_sq_mi'].str.replace('-','.')
```

```python
world.head()
```

```python
world.dtypes
```

```python
world.isnull().sum()
```

```python
world['net_migration'].fillna(world['net_migration'].median(),inplace=True)
world['infant_mortality_per_1000_births'].fillna(world['infant_mortality_per_1000_births'].median(),inplace=True)
world['gdp_$_per_capita'].fillna(world['gdp_$_per_capita'].median(),inplace=True)
world['literacy'].fillna(world['literacy'].median(),inplace=True)
world['phones_per_1000'].fillna(world['phones_per_1000'].median(),inplace=True)
world['arable'].fillna(world['arable'].median(),inplace=True)
world['crops'].fillna(world['crops'].median(),inplace=True)
world['other'].fillna(world['other'].median(),inplace=True)
world['climate'].fillna(world['climate'].median(),inplace=True)
world['birthrate'].fillna(world['birthrate'].median(),inplace=True)
world['agriculture'].fillna(world['agriculture'].median(),inplace=True)
world['industry'].fillna(world['industry'].median(),inplace=True)
world['service'].fillna(world['service'].median(),inplace=True)
world['deathrate'].fillna(world['deathrate'].median(),inplace=True)
```
